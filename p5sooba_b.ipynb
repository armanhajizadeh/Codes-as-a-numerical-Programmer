```python
import pandas as pd
import numpy as np
import datetime
import random
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import find_peaks


random.seed(5)
signup_datetime = datetime.datetime.today()
signup_datetimes = []
total_signups = 2000

for signup in range(total_signups):
    signup_datetime = signup_datetime + datetime.timedelta(days=1)
    signup_datetimes.append(signup_datetime)
    
df = pd.DataFrame({'user_created_at': signup_datetimes})
df['A'] = np.random.uniform(1, 10, df.shape[0])
df['Date'] = df['user_created_at'].dt.strftime('%Y-%m-%d')
df.drop('user_created_at', inplace=True, axis=1)


df.drop('Date', inplace=True, axis=1)
op_col = []
for i in df['A']:
    op_col.append(i)

aa = list(map(float, op_col))
bb = np.array(aa)

peaks, _ = find_peaks(bb, height=np.median(bb))
plt.plot(bb)
plt.plot(peaks, bb[peaks], "x")
plt.show

```
